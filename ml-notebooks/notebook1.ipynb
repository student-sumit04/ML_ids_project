```python
import numpy as np
import pandas as pd
import os
import glob
import joblib
import warnings

warnings.filterwarnings("ignore")


for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
csv_files = glob.glob(
    "/kaggle/input/**/*.csv",
    recursive=True
)

csv_files
```

```python
for file in csv_files:
    size = os.path.getsize(file)/(1024*1024)

    print(
        file.split("/")[-1],
        round(size,2),
        "MB"
    )
```

```python
dfs=[]


for file in csv_files:

    print("Loading:",file)

    temp=pd.read_csv(file)

    temp=temp.sample(
        n=min(100000,len(temp)),
        random_state=42
    )

    dfs.append(temp)



df=pd.concat(
    dfs,
    ignore_index=True
)


df.shape
```

```python
df.columns = df.columns.str.strip()
df.columns
```

```python
df = df[
    df["Label"]!="Label"
]
```

```python
df["Label"].value_counts()
```

```python
drop_cols=[
    "Flow ID",
    "Src IP",
    "Dst IP",
    "Src Port",
    "Timestamp"
]


df.drop(
    columns=drop_cols,
    inplace=True,
    errors="ignore"
)
```

```python
df.shape
```

```python
for col in df.columns:

    if col!="Label":

        df[col]=pd.to_numeric(
            df[col],
            errors="coerce"
        )
```

```python
df.info()
```

```python
df.replace(
    [np.inf,-np.inf],
    np.nan,
    inplace=True
)
```

```python
df.isnull().sum().sort_values(
    ascending=False
).head(10)
```

```python
df.dropna(
    inplace=True
)
```

```python
df.shape
```

```python
before=df.shape[0]

df.drop_duplicates(
    inplace=True
)

after=df.shape[0]


print(
"Removed:",
before-after
)
```

```python
from sklearn.preprocessing import LabelEncoder


encoder=LabelEncoder()


df["Label"]=encoder.fit_transform(
    df["Label"]
)
```

```python
joblib.dump(
    encoder,
    "/kaggle/working/label_encoder.pkl"
)
```

```python
for i,name in enumerate(
    encoder.classes_
):
    print(
        i,
        name
    )
```

```python
df.to_parquet(
    "/kaggle/working/ids2018_clean.parquet",
    index=False
)
```
